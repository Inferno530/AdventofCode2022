```python
with open('day3.txt') as f:
 data = [i[:-1] for i in f.readlines()]

total = 0

for word in data:
  p1 = word[0:len(word)//2]
  p2 = word[len(word)//2 if len(word)%2 == 0 else ((len(word)//2)+1):]
  common = list(set(p1)&set(p2))
  for letter in common:
    if letter.isupper():
      value = ord(letter)-38
    else:
      value = ord(letter)-96
    total = total + value


print(total)
```

```python
with open('day3.txt') as f:
 data = [i[:-1] for i in f.readlines()]

total = 0

for i in range(0, len(data), 3):
  p1 = data[i]
  p2 = data[i+1]
  p3 = data[i+2]
  common = list(set(p1)&set(p2)&set(p3))
  for letter in common:
    if letter.isupper():
      value = ord(letter)-38
    else:
      value = ord(letter)-96
    total = total + value

print(total)
```
