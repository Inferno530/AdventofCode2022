```python
#if I win with Rock: 1 point
#if I win with Paper: 2 points
#if I win with Scissors: 3 points

#if I lose: 0 points
#if I draw: 3 points
#if I win: 6 points

#A, X: Rock
#B, Y: Paper
#C, Z: Scissors
```

```python
with open('day2.txt') as f:
	data = f.readlines()

dataList = []

for line in data:
	dataList.append(line.split())
```

```python
def searchList(data):
  score = 0

  for i in data:
    if i == ['A', 'X']: #rock draw
      score = score + 1 + 3
    if i == ['B', 'Y']: #paper draw
      score = score + 2 + 3
    if i == ['C', 'Z']: #scissors draw
      score = score + 3 + 3
    if i == ['A', 'Y']: #rock, paper
      score = score + 2 + 6
    if i == ['A', 'Z']: #rock, scissors
      score = score + 0 + 3
    if i == ['B', 'X']: #paper, rock
      score = score + 0 + 1
    if i == ['B', 'Z']: #paper, scissors
      score = score + 3 + 6
    if i == ['C', 'X']: #scissors, rock
      score = score + 1 + 6
    if i == ['C', 'Y']: #scissors, paper
      score = score + 0 + 2

  return score

print(searchList(dataList))
```

```python
"""
if X, then lose
if Y, then draw
if Z, then win
"""

def again(data):
  score = 0

  for i in data:
    if i == ['A', 'X']: #rock, lose > scissor
      score = score + 0 + 3
    if i == ['B', 'Y']: #paper, draw > paper
      score = score + 3 + 2
    if i == ['C', 'Z']: #scissor, win > rock
      score = score + 6 + 1
    if i == ['A', 'Y']: #rock, draw > rock
      score = score + 3 + 1
    if i == ['A', 'Z']: #rock, win > paper
      score = score + 6 + 2
    if i == ['B', 'X']: #paper, lose > rock
      score = score + 0 + 1
    if i == ['B', 'Z']: #paper, win > scissor
      score = score + 6 + 3
    if i == ['C', 'X']: #scissor, lose > rock
      score = score + 0 + 1
    if i == ['C', 'Y']: #scissor, draw > scissor
      score = score + 3 + 3

  return score

print(again(dataList))
```
